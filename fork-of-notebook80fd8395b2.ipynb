```python

import pandas as pd


df= pd.read_csv("/kaggle/input/sales-data/sales_data_sample.csv.csv", encoding="latin1")

df["COST"] = df["PRICEEACH"] * df["QUANTITYORDERED"] * 0.7
df["PROFIT_MARGIN"] = round(((df["SALES"] - df["COST"]) / df["SALES"]) * 100, 2)

df.to_csv("../sales_data_cleaned.csv", index=False)
print("✅ Cleaned data with KPI saved to input/sales_data_cleaned.csv")
```
